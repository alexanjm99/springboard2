```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
path = 'C:/Users/u148533/heart_data_clean.csv'
dat = pd.read_csv(path)
```

```python
dat.head()
```

```python
#create dummies for categorical variables
sex = pd.get_dummies(dat['sex'], prefix = "sex")
cp = pd.get_dummies(dat['cp'], prefix = "cp")
fbs = pd.get_dummies(dat['fbs'], prefix = "fbs")
restecg = pd.get_dummies(dat['restecg'], prefix = "restecg")
exang = pd.get_dummies(dat['exang'], prefix = "exang")
slope = pd.get_dummies(dat['slope'], prefix = "slope")
ca = pd.get_dummies(dat['ca'], prefix = "ca")
thal = pd.get_dummies(dat['thal'], prefix = "thal")
```

```python
frames = [dat, sex, cp, fbs, restecg, exang, slope, ca, thal]
```

```python
#create new dataset with dummies concatenated on
dat2 = pd.concat(frames, axis = 1)
```

```python
#drop original versions of dummy variables
dat2 = dat2.drop(columns = ['sex','cp', 'fbs', 'restecg',
                                  'exang','slope','ca','thal'])

dat2.head()
```

```python
y = dat2["target"]
x = dat2.drop(["target"],axis=1).values
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```
